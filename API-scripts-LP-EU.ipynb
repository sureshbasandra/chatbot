```python
import requests

def getOrgs():
    url = "https://lo.bc-platform.liveperson.net/bot-platform-manager-0.1/organization"

    payload = ""
    headers = {
        'Authorization': AUTH_KEY,
        'OrganizationId': ORG_ID,
        'Content-Type': "application/json",
        'cache-control': "no-cache"
        }

    response = requests.request("GET", url, data=payload, headers=headers)

    return response.json()

result_getOrgs = getOrgs()
result_getOrgs
```

```python
import requests

def getAllBotsInOrg(id):
    url = "https://lo.bc-platform.liveperson.net/bot-platform-manager-0.1/chatbots"

    payload = ""
    headers = {
        'Authorization': AUTH_KEY,
        'OrganizationId': id,
        'Content-Type': "application/json",
        'cache-control': "no-cache"
        }

    response = requests.request("GET", url, data=payload, headers=headers)

    return response.json()
```

```python
import copy
all_orgs = orgs['successResult']['organziations']

bots_with_org = []
for i, org in enumerate(all_orgs):
    try:
        print(i, org)
        bots = getAllBotsInOrg(org['id'])
        for bot in bots['successResult']['chatBotSummaryList']:
            bot['orgName'] = org.get('name', None)
            bot['orgId'] = org.get('id', None)
            bot['orgCountry'] = org.get('country', None)
            bots_with_org.append(copy.deepcopy(bot))
    except:
        continue
```
